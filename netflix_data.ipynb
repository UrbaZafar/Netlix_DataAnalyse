```python
# i am gonna be using matplotlib pandas and others to analyze my netlfix data
import pandas as pd
import matplotlib.pyplot as plt

# -------using pandas-------
df=pd.read_csv("C:\\Users\\urbaz\\Downloads\\NetflixViewingHistory2.csv.csv")
# now lets explore our data
print(df.head())
# show the first 5 rows of data

print(df.info())
# shows the info abput the titles


# -----Analyze it ---------
# using matplotlib 

# a bar chart of top shows

# show me the title of the top 5 shows 
top_shows=df["Title"].value_counts().head(5) 
plt.bar(top_shows.index, top_shows.values)
# plt.bar creates a bar
plt.xlabel("Show title")
plt.ylabel("number of episodes watched")
plt.title("netflix data")
plt.show()
```
